```python
reason_1 = reasons.loc[:,1:14].max(axis=1)
reason_2 = reasons.loc[:,15:17].max(axis=1)
reason_3 = reasons.loc[:,18:22].max(axis=1)
reason_4 = reasons.loc[:,23:].max(axis=1)
```

```python
df = pd.concat([df, reason_1, reason_2, reason_3, reason_4], axis=1)
df
```
